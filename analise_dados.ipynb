```python
import re
print(train_df['all_rtts'].head(20))
def mean_str(str1):
    mean_sum=0
    n=0
    #str_num=re.split('[ |] ', str1)
    str1=str1.replace(']','')
    str1=str1.replace('[','')
    list1=str1.split(',')
    for item in list1:
        #item=item.replace(']','')
        #item=item.replace('[','')
        mean_sum+=float(item)
        n+=1
        #print(str1,' ',mean_sum,'',n)
    return mean_sum/(n+(n==0))

temp=[]
for item in train_df['all_rtts']:
    temp.append(mean_str(item))
```
