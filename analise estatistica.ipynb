```python
def robustness_analysis_hyper(errors_df, post_hoc_p_values_matrix):

    # Calcular Rank Médio (Menor MSE = Rank 1)
    ranks = errors_df.apply(lambda row: rankdata(row, method='average'), axis=1)
    mean_ranks = ranks.mean()
    
    robust_config_idx = np.argmin(mean_ranks)
    robust_config = errors_df.columns[robust_config_idx]
    print(f"Hiperparâmetro Robusto: {robust_config} (Rank Médio: {mean_ranks[robust_config_idx]:.2f})")
    
    elite_group = []
    print("\n--- Grupo de Elite (Estatística Equivalente ao Líder) ---")
    for config in errors_df.columns:
        if config == robust_config:
            elite_group.append(config)
        else:
            # Pega o p-valor entre o líder e a configuração atual
            p_val = post_hoc_p_values_matrix.loc[robust_config, config]
            if p_val >= 0.05: # Não rejeita H_0 (são iguais) com confiança de 95%
                print(f"{config} (p={p_val:.4f}) -> Empate Técnico")
                elite_group.append(config)
            else:
                print(f"{config} (p={p_val:.4f}) -> Pior que o líder")
    
    min_config = np.min([int(config) for config in elite_group])
    
    print(f"\nConfigurações finalistas: {elite_group}")
    print(f"Menor Hiperparâmetro: {min_config}")

    return str(min_config)
```

### Analise arvore sem sampling

```python
post_hoc_p_values_matrix_KNR = sp.posthoc_nemenyi_friedman(f1_tree)
post_hoc_p_values_matrix_KNR.round(2)
```

```python
hyper_tree = robustness_analysis_hyper(f1_tree, post_hoc_p_values_matrix_KNR)
```
